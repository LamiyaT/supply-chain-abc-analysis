```python
df=pd.read_csv("supply_chain_data 2.csv")
df.head()
```

```python
df=df.drop(columns=["Production volumes","Inspection results","Defect rates","Transportation modes","Routes","Supplier name","Shipping carriers","Shipping times","Supplier name"])
df.head()
```

```python
df = df.sort_values(by="Revenue generated", ascending=False)
df.head()
```

```python
total_revenue=df["Revenue generated"].sum()
total_revenue
```

```python
df['Percentage']=(df['Revenue generated']/ total_revenue ) *100
df.head()
```

```python
df['cumulative']=df['percentage'].cumsum()
df.head()
```

```python
def classify(x):
  if x <= 80:
    return 'A'
  elif x <= 95:
      return 'B'
  else :
        return 'C'
```

```python
df['Category']=df['cumulative'].apply(classify)
df
```

```python
df['Category'].value_counts()
```

```python
import matplotlib.pyplot as plt

# Select top 10 products
top10 = df.head(10)

plt.figure(figsize=(10,5))
plt.bar(top10['SKU'], top10['Revenue generated'])
plt.xticks(rotation=45)
plt.xlabel('SKU')
plt.ylabel('Revenue')
plt.title('Top 10 Products by Revenue')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(df['cumulative'], marker='o')
plt.axhline(y=80)
plt.axhline(y=95)

plt.xlabel('Number of Products')
plt.ylabel('Cumulative Percentage')
plt.title('ABC Analysis - Pareto Curve')

plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(6,4))
sns.countplot(x='Category', data=df)

plt.title('ABC Category Distribution')
plt.show()
```
